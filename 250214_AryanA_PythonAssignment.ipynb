#Question 1
The first question involves arithmetic with a small application of loops. You are given an
array of n integers, that is, a1, a2, . . . , an, and you are required to output the maximum and
minimum among them.

```python
array = [1, 2, 3, 5, 4]

def max_elem(array):
  # Direct method: max(array)
  max = array[0]
  for elem in array[1::]:
    if elem > max:
      max = elem

  return max

def min_elem(array):
  # Direct method: min(array)
  min = array[0]
  for elem in array[1::]:
    if elem < min:
      min = elem

  return min

print(f"The given array is: {array}")
print(f"Max element of array: {max_elem(array)}")
print(f"Min element of array: {min_elem(array)}")
```

#Question 2
This question is for familiarity with strings. You are needed to input a string and check if it
is a palindrome.

```python
string = "level"

def is_palindrome(string):
  # Direct method: string == string[::-1]
  # Using two pointer method
  left_ptr = 0
  right_ptr = len(string) - 1
  while left_ptr < right_ptr:
    if string[left_ptr] != string[right_ptr]:
      return False
    else:
      left_ptr += 1
      right_ptr -= 1

  return True

print(f"The given string is: {string}")
print(f"Is the given string a palindrome: {is_palindrome(string)}")
```

#Question 3
You will need a basic understanding of sorting for this question. Note that there is a direct
Python function to sort a sequence of integers. You are given an array of n integers, that is,
a1, a2, . . . , an, and a number x. You need to find a pair of elements whose absolute difference
is x. You are assured that such a pair exists in the array.

```python
array = [1, 3, 4, 7]
x = 3

def get_pair(array, x):
  # Using two-pointer method
  array.sort()

  left_ptr = 0
  right_ptr = len(array) - 1
  while left_ptr < right_ptr:
    diff = array[right_ptr] - array[left_ptr]
    if diff == x:
      return array[left_ptr], array[right_ptr]
    elif diff > x:
      left_ptr += 1
    elif diff < x:
      right_ptr -= 1

print(f"The given array is: {array}")
print(f"The given number 'x' is: {x}")
print(f"The pair of elements is: {get_pair(array, x)}")
```

#Question 4 (Optional)
You are given an array of length n, and distinct integers a, b, and c. All elements of the array
are one of a, b, or c. You are required to sort the array without the use of a conventional
sorting algorithm.

```python
array = [1, 2, 1, 3, 2, 1, 2, 1, 1, 3, 2]

def non_conventional_sort(array):
  # Find the three unique numbers in increasing order
  unique_nums = list(set(array))
  unique_nums.sort()

  small_num = unique_nums[0]
  mid_num = unique_nums[1]
  large_num = unique_nums[2]

  # Using three pointers
  small_ptr = 0
  mid_ptr = 0
  large_ptr = len(array) - 1

  while mid_ptr < large_ptr:
    if array[mid_ptr] == small_num:
      array[small_ptr], array[mid_ptr] = array[mid_ptr], array[small_ptr]
      small_ptr += 1
      mid_ptr += 1
    elif array[mid_ptr] == mid_num:
      mid_ptr += 1
    else:
      array[large_ptr], array[mid_ptr] = array[mid_ptr], array[large_ptr]
      large_ptr -= 1

  return array

print(f"The given array is: {array}")
print(f"The sorted array is: {non_conventional_sort(array)}")
```
